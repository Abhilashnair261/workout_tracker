```python

classify = predictions[["Top","Bottom"]] > 0.2

state = ""
counter = 0
for idx, (top, bottom) in classify.iterrows(): 
    if top and bottom: 
        current = ""
    elif top: 
        current = "T"
    elif bottom: 
        current = "B"
    else: 
        current = ""

    if (state=="B" or state=="TB") and current=="T":
        state = "T"
        counter += 1
    else: 
        state += current if current not in state else ""
```
